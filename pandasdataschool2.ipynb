```python
#long method for sorting the time in order
booleans=[]
for length in movies.duration:
    if length>=200:
        booleans.append(True)
    else:
        booleans.append(False)

booleans[0:5]

    
```

```python
is_long=pd.Series(booleans)
is_long.head()
```

```python
movies[is_long].head()
```

```python
#other method 
islong=movies.duration>=200
is_long.head()
```

```python
movies[islong].head()
```

```python
#third method most important 
movies[movies.duration>=200].genre.head()
```
